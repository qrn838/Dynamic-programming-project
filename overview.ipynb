```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt


%load_ext autoreload
%autoreload 2
```

```python
data = loadmat('Data/Moments_hazard_wcontrols.mat')

# Access the 'Moments' table
moments = data['Moments']

# Determine the number of elements in moments_table
num_elements = moments.shape[0]

# Calculate the number of elements to include in moments_before
num_elements_before = num_elements // 2 

# Create moments_before containing exactly half the elements in moments_table
moments_before = moments[1:num_elements_before]
moments_after = moments[num_elements_before+1:72]

# Access the 'VCcontrols' table
vc_controls = data['VCcontrol']

vc_controls_before = vc_controls[:num_elements_before, :num_elements_before]
vc_controls_after = vc_controls[num_elements_before:, num_elements_before:]



# Generate x-axis values for moments_before and moments_after
x_values_before = np.arange(len(moments_before)) * 15 + 15
x_values_after = np.arange(len(moments_after)) * 15 + 15

# Plot moments_before and moments_after on the same graph
plt.plot(x_values_before, moments_before, label='Actual hazard, Before')
plt.plot(x_values_after, moments_after, label='Actual hazard, After')

plt.xlabel('Days')
plt.ylabel('Moment Values')
plt.title('Hazard Before and After the reform')
plt.grid(True)
plt.legend()
plt.show()
```
